```
    public interface CostManage {void CalculateCost();}

    public abstract class ShippingOption : CostManage
    {
        public int DeliveryOptionID {get;set;}
        public string DeliveryOptionName {get;set;}
        public double DeliveryCost {get;set;} = 10;

        public abstract void EstimateDeliveryTime();
        void CostManage.CalculateCost() {Console.WriteLine(DeliveryCost+"$");}
        public abstract void DeliveryDetails();

        public ShippingOption(double Cost, int id, string name) 
        {DeliveryCost=Cost; DeliveryOptionID = id; DeliveryOptionName = name;}
    }

    class StandartDelivery : ShippingOption
    {
        public string AvgTime {get;set;}
        public override void EstimateDeliveryTime() {Console.WriteLine(AvgTime);}
        public override void DeliveryDetails() {Console.WriteLine("Standart Delivery Option.");}
        
        public StandartDelivery(double cost, int id, string name) : base (cost,id,name) {}
    }

    class ExpressDelivery : ShippingOption
    {
        public string MinTime {get;set;}
        public override void EstimateDeliveryTime() {Console.WriteLine(MinTime);}
        public override void DeliveryDetails() {Console.WriteLine(DeliveryOptionName+". Faster Delivery Option.");}

        public ExpressDelivery(int id,string name) : base(20,id,name) {}
        public ExpressDelivery(double Cost,  int id,string name) : base(Cost,id,name) {}
    }

public class DeliveryManager<T> where T : ShippingOption
    {
    private List<T> deliveryOptions = new List<T>();

    public void AddDeliveryOption(T option)
    {
        deliveryOptions.Add(option);
    }

    public void ProcessAllDeliveries()
    {
        foreach (var delivery in deliveryOptions)
        {
            Console.WriteLine($"Processing: {delivery.DeliveryOptionName}");
            ((CostManage)delivery).CalculateCost();
            delivery.DeliveryDetails();
            delivery.EstimateDeliveryTime();
            Console.WriteLine("---");
        }
    }
    }

    ExpressDelivery mydel = new ExpressDelivery(15,1,"A");
    StandartDelivery del2 = new StandartDelivery(10,2,"B");

    DeliveryManager<ShippingOption> Ships = new DeliveryManager<ShippingOption>();
    Ships.AddDeliveryOption(mydel);Ships.AddDeliveryOption(del2);

    Ships.ProcessAllDeliveries();
```
